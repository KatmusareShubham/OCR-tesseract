```
def ppsn(img):

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  #th,thresh = cv2.threshold(gray, 130, 255, cv2.THRESH_TRUNC)
  th,thresh = cv2.threshold(gray, 205, 255, cv2.THRESH_TRUNC)


  config = ('-l eng --oem 3 --psm 6')

  text = pytesseract.image_to_string(thresh,lang='eng') 
  print(text)
  
  user_dict = {}

  l = text.strip().replace("\n"," ").replace('|','')
  pattern = re.compile(r'\s+') 
  l = re.sub(pattern, ' ', l)
  l = l.split(' ')

  user_dict['Client_surname'] = str(" ".join(l[l.index('SURNAME')+1:l.index('TOSAINM')]))
  user_dict['Client_name'] = str(' '.join(l[l.index('FORENAME')+1:l.index('FORENAME')+3]))
  user_dict['ppsn_expiry'] = str(l[l.index('SIGNATURE')+1])

  return user_dict
```
